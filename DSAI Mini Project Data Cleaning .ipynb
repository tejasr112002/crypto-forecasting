It was observed that most cryptocurrencies do not have sufficient data on Alpha Vantage for Exploratory Analysis and Deep Learning. Thus, we shortlisted the top 10 cryptocurrencies with 1000 data points that we will use for the rest of the project, which are listed in crypto_dflist. 

```python
crypto_dflist = [BTC_data,ETH_data,BNB_data,XRP_data,ADA_data,LTC_data,
                 XLM_data,EOS_data,QTUM_data,TRX_data]
```

```python
cleaned_data = pd.DataFrame(BTC_data['timestamp'])
cleaned_data.head()
```

```python
cryptolist_ar = ['BTC','ETH','BNB','XRP','ADA','LTC','XLM','EOS','QTUM','TRX']
```

Daily Closing Price (USD) values and Daily Market Capitalization (USD) values were extracted for all 10 cryptocurrencies and compiled into a single dataframe 

```python
import itertools

for (i,j) in zip(cryptolist_ar, crypto_dflist):
    cleaned_data[i + ' close(USD)'] = j['close (USD)']
    cleaned_data[i + ' mktcap(USD)'] = j['market cap (USD)']
```
